#Reshape sales data into long format.

This allows for us to merge the sales dataset with the calendar dataset. These datasets are large, however; and any operations will cause the Python session to crash. Thus, before we can do anything else, we need to deal with the storage and memory that these datasets take up.

```python
#This code came from: https://www.kaggle.com/competitions/m5-forecasting-accuracy/data. I take no credit for it.

#Downcast in order to save memory
def downcast(df):

  '''
  Changes column types in the dataframe to reduce memory usage.
  Input:
    df (pandas dataframe): dataframe to be downcast.
  Output:
    df (pandas dataframe): dataframe with reduced memory usage.
  '''
  cols = df.dtypes.index.tolist()
  types = df.dtypes.values.tolist()
  for i,t in enumerate(types):
        if 'int' in str(t):
            if df[cols[i]].min() > np.iinfo(np.int8).min and df[cols[i]].max() < np.iinfo(np.int8).max:
                df[cols[i]] = df[cols[i]].astype(np.int8)
            elif df[cols[i]].min() > np.iinfo(np.int16).min and df[cols[i]].max() < np.iinfo(np.int16).max:
                df[cols[i]] = df[cols[i]].astype(np.int16)
            elif df[cols[i]].min() > np.iinfo(np.int32).min and df[cols[i]].max() < np.iinfo(np.int32).max:
                df[cols[i]] = df[cols[i]].astype(np.int32)
            else:
                df[cols[i]] = df[cols[i]].astype(np.int64)
        elif 'float' in str(t):
            if df[cols[i]].min() > np.finfo(np.float16).min and df[cols[i]].max() < np.finfo(np.float16).max:
                df[cols[i]] = df[cols[i]].astype(np.float16)
            elif df[cols[i]].min() > np.finfo(np.float32).min and df[cols[i]].max() < np.finfo(np.float32).max:
                df[cols[i]] = df[cols[i]].astype(np.float32)
            else:
                df[cols[i]] = df[cols[i]].astype(np.float64)
        elif t == object:
            if cols[i] == 'date':
                df[cols[i]] = pd.to_datetime(df[cols[i]], format='%Y-%m-%d')
            else:
                df[cols[i]] = df[cols[i]].astype('category')
  return df

salesVal = downcast(salesVal)
prices = downcast(prices)
cal = downcast(cal)
```

```python
#Check our datasets for stationarity
#We will use the ADF test to do this

def adfuller_test(sales_series):
    result=adfuller(sales_series)
    labels = ['ADF Test Statistic','p-value','#Lags Used','Number of Observations Used']
    for value,label in zip(result,labels):
        print(label+' : '+str(value) )
    if result[1] <= 0.05:
          print("Strong evidence against the null hypothesis(Ho), reject the null hypothesis. Data has no unit root and is stationary")
    else:
          print("Weak evidence against null hypothesis. Data is not stationary.")
    print("================================")
    print("================================")




'''
adfuller_test(salesCA['total_sale'])
adfuller_test(salesTX['total_sale'])
adfuller_test(salesWI['total_sale'])
'''
```

## Fitting the Model

The data is now stationary! We can begin fitting the model.

```python
#Set the indices in the 3 datasets to the date column
salesCA.index = pd.to_datetime(salesCA["date"])
salesTX.index = pd.to_datetime(salesTX["date"])
salesWI.index = pd.to_datetime(salesWI["date"])
```

```python
display(salesCA.head())
display(salesTX.head())
display(salesWI.head())
```

```python
#fit the ARIMA models. This may take a while.
modelCA = pm.auto_arima(salesCA['total_sale_log'], seasonal=True, m=12, trace=True)
modelTX = pm.auto_arima(salesTX['total_sale_log'], seasonal=True, m=12, trace=True)
modelWI = pm.auto_arima(salesWI['total_sale_log'], seasonal=True, m=12, trace=True)
```

We now have 3 Arima models

```python
#Forecast with these models
#Reusing code from HW3 to make these plots.

#Want to forecast the next 28 days.
modelFits = [modelCA, modelTX, modelWI]
for i, sales_df in enumerate(salesList):


  #plt.plot(i.loc["date"], i.loc["total_sale_log"])


  forecast, conf_int = modelFits[i].predict(n_periods = 28, return_conf_int=True)


  #yhat_conf_int = forecast.conf_int(alpha=0.05)

  #Create a date range for the forecast
  last_date = sales_df.index[-1]
  #Convert last_date to datetime
  last_date = pd.to_datetime(last_date)
  future_dates = pd.date_range(start=last_date + pd.DateOffset(1), periods=28)
  future_dates_str = [d.strftime('%Y-%m-%d') for d in future_dates]



  # Plot the actual values and the forecasted values
  plt.figure(figsize=(12, 6))
  # Convert 'date' column to datetime objects for proper plotting with dates
  sales_df['date'] = pd.to_datetime(sales_df['date'])
  plt.plot(sales_df["date"], sales_df["total_sale_log"], label='Actual')

  # Convert future_dates_str to datetime objects as well
  future_dates = pd.to_datetime(future_dates_str)
  plt.plot(future_dates, forecast, label='Forecast', color='red')
  plt.fill_between(future_dates, conf_int[:, 0], conf_int[:, 1], color='skyblue', alpha=0.3, label='Confidence Interval')

  plt.xlabel('Date')
  plt.ylabel('Log Transformed Total Sales')
  plt.title(f'Forecast for {stateList[i]}')
  plt.legend()
  plt.grid(True)
  plt.show()
```
